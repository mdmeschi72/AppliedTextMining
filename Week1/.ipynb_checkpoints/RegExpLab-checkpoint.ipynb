# Reg expression lab 

```python
import re
import pandas as pd
```

```python
time_sentences = ["Monday: The doctor's appointment is at 2:45pm.", 
                  "Tuesday: The dentist's appointment is at 11:30 am.",
                  "Wednesday: At 7:00pm, there is a basketball game!",
                  "Thursday: Be back home by 11:15 pm at the latest.",
                  "Friday: Take the train at 08:10 am, arrive at 09:00am."]

df = pd.DataFrame(time_sentences, columns=['text'])
df
```

```python
# figure out how many words in each sentence 

df['chars'] = df['text'].apply(str.strip).apply(len)
```

```python
df['token'] = df['text'].apply(str.strip).apply(str.split).apply(len)
```

```python
df = df.drop(columns='tokens')
```

```python
df
```

```python
# find which entries contain the word appointment 
df['contains_appointment'] = df['text'].str.contains('appointment')
```

```python
df
```

```python
# how many times a digit occurs in each string:

def parseDigits(text):
    dig = [d for d in text if re.search('\d',d)]
    return(len(dig))
```

```python
df['Num_of_digits'] = df['text'].apply(parseDigits)
```

```python
df
```

```python
df['listOfDigits'] = df['text'].apply(lambda x: re.findall('\d',x))
```

```python
df
```

```python
#find the hours and minutes 

df['hoursAndMinutes'] = df['text'].str.findall(r'(\d{1,2}:\d{1,2})')
```

```python
df['hours'] = df['text'].str.findall(r'\d{1,2}(?=:)')
```

```python
df['minutes'] = df['text'].str.findall(r'(?<=:)\d{1,2}')
```

```python
def repWeekday(text):
    
    weekday = re.findall('\w+day',text)[0]
    
    if weekday == 'Monday':
        abbr = 'Mon'
    elif weekday == 'Tuesday':
        abbr = 'Tue'
    elif weekday == 'Wednesday':
        abbr = 'Wed'
    elif weekday == 'Thursday':
        abbr = 'Thu'
    elif weekday == 'Friday':
        abbr = 'Fri'
    elif weekday == 'Saturday':
        abbr = 'Sat'
    else:
        abbr = 'Sun'
        
    outText = text.replace(weekday,abbr)
        
    return(outText)
    
```

```python
df['Day_abbr_text'] = df['text'].apply(repWeekday)
```
